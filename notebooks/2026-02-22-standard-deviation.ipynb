## Standard Deviation

The *Standard Deviation* (SD, Std Dev) is a measure of the amount of variation of the values of a variable about its average. A low standard deviation indicates that the values tend to be close to their average, while a high standard deviation indicates that the values are spread out over a wider range. The standard deviation is the square root of the variance. In contrast to the variance the standard deviation is directly interpretable. The variance of a dataset is expressed in squared units, whereas, the standard deviation is expressed in original units.

---

Notation: $\sigma$ or $s$

Population standard deviation: $\sigma=\sqrt{\sigma^2}=\sqrt{\frac{1}{N}\sum_{i=1}^{N}{(X_i-\mu)^2}}$

Sample standard deviation: $s=\sqrt{s^2}=\sqrt{\frac{1}{n-1}\sum_{i=1}^{n}{(X_i-\bar{X})^2}}$

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x_0 = np.random.normal(0, 1, 1000)
x_1 = np.random.normal(0, 3, 1000)
sns.kdeplot(x_0, label=f"SD: {np.std(x_0):.2f}")
sns.kdeplot(x_1, label=f"SD: {np.std(x_1):.2f}")
plt.axvline(0, c="black", label="mean")
plt.legend()
```

### Calculate Std Dev with Numpy

```python
import numpy as np

def std_dev(x, sample=False):
    n = len(x)
    if sample:
        n -= 1
    var_x = np.sum(np.pow(x - np.mean(x), 2)) / n
    return np.sqrt(var_x)

x = np.linspace(1,10,10)
std_dev(x), std_dev(x, sample=True)
```
